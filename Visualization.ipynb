###### Importing libraries essential for data visualization

```python
#MATPLOTLIB
import matplotlib.pyplot as plt
%matplotlib inline
#SEABORN
import seaborn as sns
```

## Plots for continuous variables' analysis

### ENUMERATIVE PLOTS

###### UNIVARIATE SCATTER PLOT

```python
plt.scatter(df.index,df['trestbps'])
plt.show()
```

```python
plt.scatter(df.index,df['fbs'])
plt.show()
```

###### STRIP PLOT

```python
sns.stripplot(y=df['trestbps'])
```

```python
sns.stripplot(y=df['fbs'])
```

###### SWARM PLOT

```python
#Setting figure size
sns.set(rc={'figure.figsize':(5,5)})
#Swarm-plot
sns.swarmplot(x=df['trestbps'])
```

```python
#Setting figure size
sns.set(rc={'figure.figsize':(5,5)})
#Swarm-plot
sns.swarmplot(x=df['fbs'])
```

### SUMMARY PLOTS

###### HISTOGRAM

```python
plt.hist(df['trestbps'])
```

```python
plt.hist(df['fbs'])
```

```python
sns.distplot(df['trestbps'],kde=False,color='black',bins=10)
```

```python
sns.distplot(df['fbs'],kde=False,color='black',bins=10)
```

###### DENSITY PLOT

```python
plt.figure(figsize=(5,5))
df['trestbps'].plot(kind='density')
```

```python
plt.figure(figsize=(5,5))
df['fbs'].plot(kind='density')
```

```python
from scipy import stats
import numpy as np
kdf=df['trestbps'].to_numpy()
rdf=np.hstack(kdf)
density = stats.kde.gaussian_kde(rdf)
x = np.arange(3,9,0.1)
plt.plot(x, density(x))
plt.plot(rdf,[0.01]*len(rdf), '|')
```

```python
sns.distplot(df['trestbps'],rug=True,hist=False)
```

```python
sns.distplot(df['fbs'],rug=True,hist=False)
```
