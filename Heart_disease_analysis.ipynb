```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
for x in heart.columns.tolist():
    plt.figure(figsize=(5,5))
    sns.countplot(data=heart,x=heart[x])
    plt.xticks(color='red',rotation = 90,size=10)
    plt.yticks(color='red',size=10)
    plt.title('Frequency of {}'.format(x))
    plt.xlabel(x,color='red')
    plt.ylabel('count',color='red')
    plt.show()
```

```python
for x in heart.columns.tolist():
    plt.figure(figsize=(10,8))
    sns.distplot(heart[x],color='red')
    plt.xticks(color='black',rotation = 90,size=10)
    plt.yticks(color='black',size=10)
    plt.title('Distribution of {}'.format(x),color='red')
    plt.xlabel(x,color='black')
    plt.ylabel('count',color='black')
    plt.show()
```

Relation between different columns

```python
plt.figure(figsize=(14, 14))
sns.set(font_scale=1.5)
sns.heatmap(heart.corr(), cmap='coolwarm', annot=True, annot_kws={'size':15})
plt.show()
```

Distribution of data in dataset

```python
for x in ['cp','slope']:
    print('Distribution of {}'.format(x))
    labels=heart[x].value_counts().to_frame().index.tolist()
    sizes=heart[x].value_counts().to_frame()[x].tolist()
    plt.pie(sizes,labels=labels,autopct='%0.2f%%', shadow=True)
    plt.axis('equal')
    plt.show()
```

```python
for columns in heart:
    #pd.crosstab(heart[columns],heart.target).plot(kind='bar',color=['red','green'],figsize=(10,8),label=['Disease','No-disease'])
    plt.figure(figsize=(10,5))
    sns.countplot(x=heart[columns],hue= heart['target'])
    plt.xticks(rotation = 90)
    plt.xlabel(columns,color='red',size=20)
    plt.ylabel('Number of patients',color='red',size=20)
    plt.title('Number of patients vs {}'.format(columns),size=20)
    plt.xticks(color='red',size=10)
    plt.yticks(color='red',size=10)
    plt.legend()
    plt.show()
```
